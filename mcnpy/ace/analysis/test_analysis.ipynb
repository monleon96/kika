# Testing ACE Cross Section Plotting

This notebook demonstrates the simplified `plot_cross_sections` function from the `cross_section_plotting.py` module for comparing nuclear cross sections from ACE files.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Import the simplified module
from mcnpy.ace.analysis.cross_section_plotting import plot_cross_sections
```

## Setting File Paths

Define the paths to your ACE files below:

```python
# Replace these with paths to your ACE files
ace_file1 = "../files/260560_80.02c"  # e.g., "/home/data/u235.ace"
ace_file2 = "../files/922350_71.02c"  # e.g., "/home/data/u238.ace"

# Check if files exist
for file in [ace_file1, ace_file2]:
    if not os.path.exists(file):
        print(f"Warning: {file} does not exist. Please update the path.")
```

## Plotting Cross Sections

Let's plot some important cross sections from the ACE files:

```python
# Define the ACE files and custom labels
ace_files = [ace_file1, ace_file2]
labels = ["First File", "Second File"]  # Custom labels for the legend

# Plot total cross section (MT=1)
fig1 = plot_cross_sections(
    ace_files,
    mt_number=1,  # Total cross section
    labels=labels,
    energy_range=(1e-11, 20.0),
    figsize=(10, 6)
)
plt.show()
```

```python
# Plot elastic scattering cross section (MT=2)
fig2 = plot_cross_sections(
    ace_files,
    mt_number=2,  # Elastic scattering
    labels=labels,
    energy_range=(1e-11, 20.0),
    figsize=(10, 6)
)
plt.show()
```

```python
# Plot capture cross section (MT=102)
fig3 = plot_cross_sections(
    ace_files,
    mt_number=102,  # Radiative capture (n,gamma)
    labels=labels,
    energy_range=(1e-11, 20.0),
    figsize=(10, 6)
)
plt.show()
```

## Try a Different MT Number

You can try plotting different MT numbers by changing the value below:

```python
# Try a different MT number
custom_mt = 18  # (n,alpha) reaction - change to any MT number you want to plot

fig4 = plot_cross_sections(
    ace_files,
    mt_number=custom_mt,
    labels=labels,
    energy_range=(1e-11, 20.0),
    figsize=(10, 6)
)
plt.show()
```
