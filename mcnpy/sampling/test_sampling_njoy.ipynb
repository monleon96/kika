```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.csgraph as cs

def mt_blocks(cov_nd, mt_numbers, n_groups, tol=1e-12):
    """
    Devuelve una lista de arrays con los índices (fila/columna) que forman
    cada bloque independiente MT-completo.
    """
    import numpy as np
    import scipy.sparse as sp, scipy.sparse.csgraph as cs

    n_mt  = len(mt_numbers)
    first = {k: k*n_groups for k in range(n_mt)}  # MT → columna inicial

    # --- matriz de adyacencia entre MTs ---------------------------------
    adj = np.zeros((n_mt, n_mt), dtype=bool)
    for i in range(n_mt):
        for j in range(i+1, n_mt):
            sli = slice(first[i], first[i]+n_groups)
            slj = slice(first[j], first[j]+n_groups)
            if np.any(np.abs(cov_nd[sli, slj]) > tol):
                adj[i, j] = adj[j, i] = True

    n_comp, lbl = cs.connected_components(sp.csr_matrix(adj))
    blocks = [np.concatenate(
                [np.arange(first[m], first[m]+n_groups) for m in np.where(lbl==k)[0]]
              ) for k in range(n_comp)]
    return blocks



import numpy as np, scipy.stats
from numpy.linalg import norm
from scipy.stats import qmc, norm as gauss
from mcnpy.sampling.decomposition import (
        svd_decomposition, cholesky_decomposition, eigen_decomposition)

def empirical_cov(X):
    Xc = X - X.mean(axis=0, keepdims=True)
    return (Xc.T @ Xc) / (X.shape[0] - 1)

def _decompose(cov, how):
    h = how.lower()
    if h == "svd":       return svd_decomposition(cov)
    if h == "cholesky":  return cholesky_decomposition(cov)
    if h == "eigen":     return eigen_decomposition(cov)
    raise ValueError("decomposition_method must be svd | cholesky | eigen")

def _uncorrelated(block_dim, n, method, seed, fast_forward=0):
    m = method.lower()
    if m == "random":
        rng = np.random.default_rng(seed)
        return rng.standard_normal((n, block_dim))

    if m == "lhs":
        U = qmc.LatinHypercube(d=block_dim, seed=seed).random(n)
        return scipy.stats.norm.ppf(U)

    if m == "sobol":
        sob = qmc.Sobol(d=block_dim, scramble=True, seed=seed)
        if fast_forward:
            sob.fast_forward(fast_forward)
        U = sob.random(n)
        return scipy.stats.norm.ppf(U)

    raise ValueError("sampling_method must be random | lhs | sobol")

def generate_samples_blocked(
    cov_nd: np.ndarray,
    blocks: list[np.ndarray],
    n_samples: int,
    decomposition_method: str = "svd",
    sampling_method: str = "sobol",
    seed: int | None = None,
) -> np.ndarray:
    """
    Devuelve un array (n_samples × p) de factores correlacionados,
    aplicando la descomposición por bloques pero usando un único
    generador Sobol de dimensión total para evitar covarianzas
    artificiales entre bloques.
    """
    p   = cov_nd.shape[0]
    out = np.empty((n_samples, p))

    # ─── 1. muestras no correlacionadas (una sola vez) ──────────────
    if sampling_method.lower() == "sobol":
        sob       = qmc.Sobol(d=p, scramble=True, seed=seed)
        Z_global  = scipy.stats.norm.ppf(sob.random(n_samples))
    else:
        Z_global = None    # se creará bloque a bloque

    # ─── 2. recorrer los bloques ────────────────────────────────────
    for idx in blocks:
        dim = len(idx)

        # seleccionar las columnas de Z_global o generar bloque a bloque
        if sampling_method.lower() == "sobol":
            z_b = Z_global[:, idx]     # coherencia total del net
        else:
            # para random/lhs usamos semillas derivadas
            s_b = (seed or 0) + hash(tuple(idx)) % (2**31 - 1)
            z_b = _uncorrelated(dim, n_samples, sampling_method, s_b)

        # descomposición y transformación local
        L_b = _decompose(cov_nd[np.ix_(idx, idx)], decomposition_method)
        out[:, idx] = z_b @ L_b.T

    # desplazamiento +1.0 como en tu versión original
    return out + 1.0
```
