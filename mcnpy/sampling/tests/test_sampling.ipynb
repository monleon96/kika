```python
import os

def find_negative_perturbations(zaids, sample_nums, base_dir):
    """
    Scan summary files under base_dir for given ZAID and SAMPLE_NUM lists,
    and return a list of (zaid, sample_num) pairs where a negative perturbation
    factor was found.
    """
    negatives = []
    for zaid in zaids:
        zaid_str = str(zaid)
        for sample_num in sample_nums:
            # format SAMPLE_NUM as 4-digit string, padded with zeros on the left
            sample_str = f"{sample_num:04d}"
            # build filename and full path
            filename = f"{zaid_str}0_80_{sample_str}_summary.txt"
            file_path = os.path.join(base_dir, zaid_str, sample_str, filename)

            if not os.path.isfile(file_path):
                print(f"Warning: file not found: {file_path}")
                continue

            with open(file_path, 'r') as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) < 4:
                        continue
                    try:
                        factor = float(parts[3])
                    except ValueError:
                        # skip lines that don't parse
                        continue
                    if factor < 0:
                        negatives.append((zaid, sample_num))
                        # stop checking this file once a negative is found
                        break
    return negatives
```

```python
base_dir = "/home/MONLEON-JUAN/MCNPy/mcnpy/sampling/new_perturbed_ace_files"

zaids = [25055]
sample_nums = list(range(1, 1025))  # e.g. samples 1 through 10

results = find_negative_perturbations(zaids, sample_nums, base_dir)

print("\nSummary of negative perturbation factors found:")
if results:
    for zaid, sample in results:
        print(f"  • ZAID = {zaid}, SAMPLE_NUM = {sample}")
else:
    print("  None found.")
```
