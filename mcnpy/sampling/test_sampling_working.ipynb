```python
# run_my_perturb.py

import mcnpy
from mcnpy.sampling.perturb_ace_multi import parallel_perturb
from mcnpy.cov.parse_covmat import read_scale_covmat

# 1) Prepare inputs just like you did before
scale_file_path = '../cov/files/scale.rev05.44groupcov.txt'

acefile1 = '../ace/files/260560_80.02c'
acefile2 = '../ace/files/10010_71.02c'
xsdir_file = '/soft_snc/lib/xsdir/xsdir80'

covmat = read_scale_covmat(scale_file_path)
acelist    = [acefile1, acefile2]
covmatlist = [covmat]*len(acelist)

# Perturbation parameters
mt_numbers     = [2, 4, 102, 103, 107]
energy_grid    = mcnpy.energy_grids.SCALE44
num_samples    = 2
output_dir     = "./new_perturbed_ace_files"
seed           = 42
nprocs         = 1
```

```python
# 2) Call the new parallel_perturb
print(f"Generating {num_samples} perturbed ACE files...")
parallel_perturb(
    ace_file_path      = acelist,
    mt_numbers         = mt_numbers,
    energy_grid        = energy_grid,
    covmat             = covmatlist,
    num_samples        = num_samples,
    decomposition_method="svd",
    sampling_method    ="sobol",
    output_dir         = output_dir,
    xsdir              = xsdir_file,
    seed               = seed,
    verbose            = True,
    nprocs             = nprocs,     
)
```
