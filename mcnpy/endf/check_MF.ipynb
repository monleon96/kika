```python
from mcnpy.endf.read_endf import read_endf
dir_jeff33 = "/mnt/c/Users/MONLEON-DE-LA-JAN/Documents/jndl40"
import os
import gc

# List to store filenames containing MF34
files_with_mf34 = []

# Traverse through all files in directory
processed_files = 0
for root, _, files in os.walk(dir_jeff33):
    for file in files:
        file_path = os.path.join(root, file)
        
        # Skip directories
        if os.path.isdir(file_path):
            continue
        
        processed_files += 1
        
        # Print progress every 20 files
        if processed_files % 20 == 0:
            print(f"Processed {processed_files} files")
        
        try:
            # Try to read the ENDF file
            endf_mf1 = read_endf(file_path, mf_numbers=1)
            mt451 = endf_mf1.mf[1].mt[451]
            directory_df = mt451.directory

            # Check if MF34 exists in this file
            if 34 in directory_df['MF'].values:
                files_with_mf34.append(file_path)
            
            # Explicitly release the ENDF6 object to free memory
            del endf_mf1
            gc.collect()
        except Exception as e:
            # File might not be an ENDF file or could have other issues
            pass

# Print the list of files containing MF34
print("\nFiles containing MF34:")
for file in files_with_mf34:
    print(file)
print(f"\nTotal files with MF34: {len(files_with_mf34)} out of {processed_files} processed")
```
