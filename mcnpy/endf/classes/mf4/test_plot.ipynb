```python
# Print the type and size of each element in the list
coeffs = endf.mf[4].mt[2].legendre_coefficients
print("Type:", type(coeffs))
for i, elem in enumerate(coeffs):
    if hasattr(elem, 'shape'):
        print(f"Element {i}: shape = {elem.shape}")
    elif hasattr(elem, '__len__'):
        print(f"Element {i}: length = {len(elem)}")
    else:
        print(f"Element {i}: type = {type(elem)}")
```
