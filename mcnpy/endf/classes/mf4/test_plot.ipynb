```python
# Test all three improvements
import sys
import importlib

# Remove modules from cache to force fresh imports
modules_to_reload = [
    'mcnpy.endf.parsers.parse_endf',
    'mcnpy.endf.classes.endf', 
    'mcnpy.endf.classes.mf4.plotting'
]

for module_name in modules_to_reload:
    if module_name in sys.modules:
        del sys.modules[module_name]

# Fresh import and parsing
from mcnpy.endf.parsers.parse_endf import parse_endf_file
endf_improved = parse_endf_file(Fe56_ednf)

print("=== Testing Improvements ===")
print(f"1. MAT number: {endf_improved.mat}")
print(f"2. ZAID (now as property): {endf_improved.zaid}")  
print(f"3. Isotope symbol: {endf_improved.get_isotope_symbol()}")

# Test the improved plotting function
from mcnpy.endf.classes.mf4.plotting import plot_legendre_coefficients_from_endf

print("\n=== Creating plots with improvements ===")

# Plot with uncertainties
print("Plot WITH uncertainties:")
fig_with_unc = plot_legendre_coefficients_from_endf(
    endf_improved, 
    mt=2, 
    orders=[1, 2, 3], 
    include_uncertainties=True,
    figsize=(12, 7)
)
fig_with_unc.show()

print("\nPlot WITHOUT uncertainties:")
# Plot without uncertainties  
fig_without_unc = plot_legendre_coefficients_from_endf(
    endf_improved, 
    mt=2, 
    orders=[1, 2, 3], 
    include_uncertainties=False,
    figsize=(12, 7)
)
fig_without_unc.show()

print("\n=== Improvements Summary ===")
print(" ZAID is now a computed property (not stored attribute)")
print(" Title is more concise without 'vs Energy'") 
print(" Legend uses proper mathematical subscripts: a, a, a")
```
