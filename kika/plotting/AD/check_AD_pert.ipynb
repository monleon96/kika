```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the parquet file
file_path = '/share_snc/snc/JuanMonleon/plot_samples_leg/endf_perturbation_factors_20250920_084226.parquet'
df = pd.read_parquet(file_path)

# Display basic information about the data
print("Shape:", df.shape)
print("\nColumn names (first 30):")
print(df.columns.tolist()[:30])
print("\nFirst few rows:")
df.head()
```

```python
# Analyze the column structure to understand Legendre orders
# Extract all unique Legendre orders from column names
import re

legendre_orders = set()
for col in df.columns:
    if col != 'Sample_ID':
        # Extract the Legendre order (L number) from column name
        match = re.search(r'_L(\d+)_', col)
        if match:
            legendre_orders.add(int(match.group(1)))

print("Legendre orders found:", sorted(legendre_orders))

# Count columns for each Legendre order
for order in sorted(legendre_orders):
    cols = [col for col in df.columns if f'_L{order}_' in col]
    print(f"\nOrder L{order}: {len(cols)} columns")
    if order == 6:
        print("Sample column names for L6:")
        print(cols[:5])
```

```python
# Extract all L6 columns
l6_columns = [col for col in df.columns if '_L6_' in col]
print(f"Total L6 columns: {len(l6_columns)}\n")

# Display all L6 column names to see the energy bins
for i, col in enumerate(l6_columns, 1):
    print(f"{i}. {col}")
```

```python
# Calculate variance and standard deviation for each L6 energy bin
# Create a results dataframe
results = []

for col in l6_columns:
    # Extract energy range from column name
    energy_range = col.replace('Fe56_MT2_L6_', '')
    
    # Get the perturbation factors for this energy bin
    perturbations = df[col].values
    
    # Calculate statistics
    mean = np.mean(perturbations)
    variance = np.var(perturbations, ddof=1)  # Sample variance (ddof=1)
    std_dev = np.std(perturbations, ddof=1)   # Sample standard deviation
    
    # Calculate relative uncertainty (std_dev / mean) if mean is not zero
    rel_uncertainty = (std_dev / mean * 100) if mean != 0 else 0
    
    results.append({
        'Energy_Bin': energy_range,
        'Mean': mean,
        'Variance': variance,
        'Std_Dev': std_dev,
        'Rel_Uncertainty_%': rel_uncertainty,
        'N_Samples': len(perturbations)
    })

results_df = pd.DataFrame(results)
print("Statistics for Order 6 Legendre Coefficients by Energy Bin:")
print("=" * 80)
results_df
```

```python
# Create a clean summary showing just Energy Bin and Std_Dev (uncertainty)
# This is what you need to compare with the original covariance file
uncertainty_summary = results_df[['Energy_Bin', 'Std_Dev', 'Rel_Uncertainty_%']].copy()
uncertainty_summary.columns = ['Energy_Bin', 'Uncertainty_from_Distribution', 'Relative_Uncertainty_%']

print("\nUncertainty extracted from L6 perturbation distributions:")
print("=" * 80)
print(uncertainty_summary.to_string(index=False))
print("\n" + "=" * 80)
print(f"\nTotal number of energy bins for L6: {len(uncertainty_summary)}")
print(f"Number of samples used: {results_df['N_Samples'].iloc[0]}")
```

## Interpretation

The table above shows the **uncertainty (standard deviation)** extracted from the perturbation factor distributions for each energy bin of the Order 6 Legendre coefficients.

**Key observations:**
- We have **13 energy bins** for L6 coefficients
- Analyzed **1024 samples** from the perturbation distribution
- The standard deviation represents the uncertainty in the perturbation factors

**To compare with the original covariance file:**

- Large relative uncertainties (e.g., 1873% for 2.8-3.0 MeV) indicate significant variations in the sampled perturbation factors
- Very small uncertainties (e.g., 0.000019% for low/high energies) suggest the perturbations are tightly clustered around the mean

## Compare Perturbation Uncertainties with ENDF MF34 Data

Now we'll compare the uncertainties extracted from the perturbation distributions with the uncertainties stored in the ENDF file (MF34). If the sampling was done correctly, these should agree closely.

```python
# Load the ENDF file
from kika.endf.read_endf import read_endf

endf_file_path = '/share_snc/snc/JuanMonleon/jeff40_with_MF4_from_jeff33/26-Fe-56g.txt'
print(f"Loading ENDF file: {endf_file_path}")
endf_data = read_endf(endf_file_path)
print("âœ“ ENDF file loaded successfully")

# Extract MF34 covariance data for MT=2 (elastic scattering)
mt_elastic = 2
zaid = 26056

try:
    mf34_covmat = endf_data.mf[34].mt[mt_elastic].to_ang_covmat()
    print(f"\nâœ“ MF34 covariance data extracted")
    print(f"  Number of matrices: {mf34_covmat.num_matrices}")
    print(f"  Available Legendre orders: {sorted(mf34_covmat.legendre_indices)}")
except Exception as e:
    print(f"âš  Error extracting MF34 data: {e}")
    mf34_covmat = None
```

```python
# Function to extract ENDF uncertainties for a specific Legendre order
def get_endf_uncertainties_for_order(mf34_covmat, zaid, mt, order):
    """
    Extract uncertainties from ENDF MF34 covariance data for a specific Legendre order.
    
    Returns
    -------
    dict or None
        Dictionary with 'energies' and 'uncertainties' arrays, or None if not found
    """
    if mf34_covmat is None:
        return None
    
    try:
        unc_data = mf34_covmat.get_uncertainties_for_legendre_coefficient(zaid, mt, order)
        return unc_data
    except Exception as e:
        print(f"Warning: Could not get uncertainties for L={order}: {e}")
        return None


# Extract ENDF uncertainties for L5 and L6
print("\nExtracting ENDF uncertainties...")
endf_unc_l5 = get_endf_uncertainties_for_order(mf34_covmat, zaid, mt_elastic, 5)
endf_unc_l6 = get_endf_uncertainties_for_order(mf34_covmat, zaid, mt_elastic, 6)

if endf_unc_l5:
    print(f"âœ“ L5 uncertainties: {len(endf_unc_l5['uncertainties'])} energy bins")
    print(f"  Energy range: {endf_unc_l5['energies'][0]:.2e} - {endf_unc_l5['energies'][-1]:.2e} eV")
    
if endf_unc_l6:
    print(f"âœ“ L6 uncertainties: {len(endf_unc_l6['uncertainties'])} energy bins")
    print(f"  Energy range: {endf_unc_l6['energies'][0]:.2e} - {endf_unc_l6['energies'][-1]:.2e} eV")
```

```python
# Process perturbation data for L5 and L6
def process_perturbation_data_for_order(df, order):
    """
    Process perturbation data for a specific Legendre order.
    
    Returns
    -------
    pd.DataFrame
        DataFrame with columns: Energy_Bin, Mean, Std_Dev, Rel_Uncertainty_%
    """
    # Get all columns for this order
    order_columns = [col for col in df.columns if f'_L{order}_' in col]
    
    if len(order_columns) == 0:
        print(f"Warning: No columns found for L{order}")
        return None
    
    results = []
    for col in order_columns:
        # Extract energy range from column name
        energy_range = col.replace(f'Fe56_MT2_L{order}_', '')
        
        # Get the perturbation factors for this energy bin
        perturbations = df[col].values
        
        # Calculate statistics
        mean = np.mean(perturbations)
        std_dev = np.std(perturbations, ddof=1)   # Sample standard deviation
        rel_uncertainty = (std_dev / mean * 100) if mean != 0 else 0
        
        results.append({
            'Energy_Bin': energy_range,
            'Mean': mean,
            'Std_Dev': std_dev,
            'Rel_Uncertainty_%': rel_uncertainty,
            'N_Samples': len(perturbations)
        })
    
    return pd.DataFrame(results)


# Extract perturbation uncertainties for L5 and L6
print("\nProcessing perturbation data...")
pert_results_l5 = process_perturbation_data_for_order(df, 5)
pert_results_l6 = process_perturbation_data_for_order(df, 6)

if pert_results_l5 is not None:
    print(f"âœ“ L5 perturbations: {len(pert_results_l5)} energy bins")
if pert_results_l6 is not None:
    print(f"âœ“ L6 perturbations: {len(pert_results_l6)} energy bins")
```

```python
# Function to parse energy bin strings and get midpoint energies
def parse_energy_bin(energy_bin_str):
    """
    Parse energy bin string like '1.00e+05_1.50e+05' and return midpoint in eV.
    
    Returns
    -------
    float
        Midpoint energy in eV
    """
    parts = energy_bin_str.split('_')
    if len(parts) == 2:
        e_min = float(parts[0])
        e_max = float(parts[1])
        return (e_min + e_max) / 2.0
    else:
        # If format is different, try to extract numbers
        import re
        numbers = re.findall(r'[\d.]+e[+-]?\d+', energy_bin_str)
        if len(numbers) >= 2:
            return (float(numbers[0]) + float(numbers[1])) / 2.0
        return None


# Add midpoint energies to perturbation results
if pert_results_l5 is not None:
    pert_results_l5['Energy_MidPoint_eV'] = pert_results_l5['Energy_Bin'].apply(parse_energy_bin)
    
if pert_results_l6 is not None:
    pert_results_l6['Energy_MidPoint_eV'] = pert_results_l6['Energy_Bin'].apply(parse_energy_bin)

print("\nPerturbation data with energy midpoints:")
print("\nL6 sample:")
if pert_results_l6 is not None:
    print(pert_results_l6[['Energy_Bin', 'Energy_MidPoint_eV', 'Std_Dev']].head())
```

```python
# Comparison function
def compare_uncertainties(endf_unc, pert_results, order):
    """
    Compare ENDF uncertainties with perturbation-derived uncertainties.
    
    Parameters
    ----------
    endf_unc : dict
        Dictionary with 'energies' and 'uncertainties' from ENDF MF34
    pert_results : pd.DataFrame
        DataFrame with perturbation results
    order : int
        Legendre order
    
    Returns
    -------
    pd.DataFrame
        Comparison DataFrame
    """
    if endf_unc is None or pert_results is None:
        print(f"Cannot compare L{order}: missing data")
        return None
    
    # Get ENDF data
    endf_energies = endf_unc['energies']  # in eV
    endf_uncertainties = endf_unc['uncertainties']  # absolute uncertainties
    
    # Get perturbation data
    pert_energies = pert_results['Energy_MidPoint_eV'].values
    pert_uncertainties = pert_results['Std_Dev'].values
    
    # Create comparison DataFrame
    comparison_data = []
    
    # For each perturbation energy bin, find the closest ENDF energy
    for i, pert_energy in enumerate(pert_energies):
        if pert_energy is None:
            continue
        
        # Find closest ENDF energy point
        idx_closest = np.argmin(np.abs(endf_energies - pert_energy))
        endf_energy = endf_energies[idx_closest]
        endf_unc = endf_uncertainties[idx_closest]
        pert_unc = pert_uncertainties[i]
        
        # Calculate relative difference
        if endf_unc != 0:
            rel_diff_pct = ((pert_unc - endf_unc) / endf_unc) * 100
        else:
            rel_diff_pct = np.nan
        
        comparison_data.append({
            'Perturbation_Energy_eV': pert_energy,
            'ENDF_Energy_eV': endf_energy,
            'Energy_Diff_eV': abs(pert_energy - endf_energy),
            'Perturbation_Unc': pert_unc,
            'ENDF_Unc': endf_unc,
            'Absolute_Diff': abs(pert_unc - endf_unc),
            'Relative_Diff_%': rel_diff_pct,
            'Energy_Bin': pert_results.iloc[i]['Energy_Bin']
        })
    
    return pd.DataFrame(comparison_data)


# Compare uncertainties for L5 and L6
print("\n" + "="*80)
print("COMPARING PERTURBATION vs ENDF UNCERTAINTIES")
print("="*80)

comparison_l5 = compare_uncertainties(endf_unc_l5, pert_results_l5, 5)
comparison_l6 = compare_uncertainties(endf_unc_l6, pert_results_l6, 6)
```

```python
# Display comparison results for L5
if comparison_l5 is not None:
    print("\n" + "="*80)
    print("L5 COMPARISON: Perturbation vs ENDF Uncertainties")
    print("="*80)
    print(comparison_l5.to_string(index=False))
    print("\n" + "-"*80)
    print("SUMMARY STATISTICS FOR L5:")
    print("-"*80)
    print(f"Number of energy bins compared: {len(comparison_l5)}")
    print(f"Mean absolute difference: {comparison_l5['Absolute_Diff'].mean():.6e}")
    print(f"Mean relative difference: {comparison_l5['Relative_Diff_%'].mean():.2f}%")
    print(f"Max absolute difference: {comparison_l5['Absolute_Diff'].max():.6e}")
    print(f"Max relative difference: {comparison_l5['Relative_Diff_%'].max():.2f}%")
else:
    print("\nL5 comparison not available")
```

## Interpretation and Validation

### Expected Results:
If the perturbation sampling was done correctly, the uncertainties extracted from the perturbation distributions should closely match the uncertainties stored in the ENDF MF34 file.

### What to look for:
1. **Absolute Uncertainties**: The two lines (ENDF vs Perturbations) should overlap or be very close
2. **Relative Difference**: Should be small (ideally < 5-10%) for most energy bins
3. **Large differences** could indicate:
   - Insufficient number of samples (need more than 1024?)
   - Issues with the sampling methodology
   - Energy bin misalignment
   - Numerical precision problems

### Statistical Considerations:
- With 1024 samples, the standard error of the standard deviation is approximately Ïƒ/âˆš(2N) â‰ˆ 2.2% of the true value
- So some discrepancy (a few percent) is expected due to finite sampling
- Systematic biases (all perturbation uncertainties consistently higher/lower than ENDF) would indicate a problem with the sampling process
