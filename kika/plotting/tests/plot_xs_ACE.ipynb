```python
import kika
from kika.plotting import PlotBuilder
```

```python
ace_file_path_jeff40 = '/soft_snc/lib/ace/40/600/260560_40.06c'
ace_file_path_endfb81 = '/soft_snc/lib/ace/81/600/260560_81.06c'
```

```python
ace_jeff40 = kika.read_ace(ace_file_path_jeff40)
ace_endfb81 = kika.read_ace(ace_file_path_endfb81)
```

```python
xs_data_jeff40 = ace_jeff40.to_plot_data('cross_section', mt=2, label='JEFF-4.0')
xs_data_b81 = ace_endfb81.to_plot_data('cross_section', mt=2, label='ENDF/B-VIII.1')
```

```python
xs_data_jeff40
```

```python
# Build the plot with log scales
fig = (PlotBuilder()
       .add_data(xs_data_jeff40)
       .add_data(xs_data_b81, linestyle='dashed')
       .set_labels(
           title='Elastic Scattering Cross Section Comparison',
           x_label='Energy (MeV)',
           y_label='Cross Section (barns)'
       )
       .set_scales(log_x=True, log_y=True)
       .build())
```
