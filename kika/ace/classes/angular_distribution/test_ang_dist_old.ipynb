```python
# Extract the actual Legendre coefficient data to compare numerically
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the energy range we're interested in
energy_min, energy_max = 1e6, 2e6
mt = 2  # elastic scattering
order = 1  # first order Legendre coefficient

print("Extracting Legendre coefficient data...")
print(f"MT={mt}, Order={order}, Energy range: {energy_min:.1e} - {energy_max:.1e} eV")

def extract_legendre_order(mf4_mt_data, order):
    """Extract Legendre coefficients for a specific order"""
    energies = np.array(mf4_mt_data.legendre_energies)
    legendre_coeffs_all = mf4_mt_data.legendre_coefficients
    
    # Extract the specific order from each energy point
    coeffs_order = []
    for i, coeffs_at_energy in enumerate(legendre_coeffs_all):
        if len(coeffs_at_energy) > order:
            coeffs_order.append(coeffs_at_energy[order])
        else:
            coeffs_order.append(0.0)  # If order not available, assume 0
    
    return energies, np.array(coeffs_order)

# Extract data from all three files
energies_ref, coeffs_ref = extract_legendre_order(endf.mf[4].mt[mt], order)
energies_s1, coeffs_s1 = extract_legendre_order(endf1.mf[4].mt[mt], order)
energies_s2, coeffs_s2 = extract_legendre_order(endf2.mf[4].mt[mt], order)

print(f"Reference: {len(energies_ref)} energy points")
print(f"Sample 1: {len(energies_s1)} energy points")
print(f"Sample 2: {len(energies_s2)} energy points")

# Filter data to our energy range of interest
energy_mask_ref = (energies_ref >= energy_min) & (energies_ref <= energy_max)
energy_mask_s1 = (energies_s1 >= energy_min) & (energies_s1 <= energy_max)
energy_mask_s2 = (energies_s2 >= energy_min) & (energies_s2 <= energy_max)

energies_ref_filtered = energies_ref[energy_mask_ref]
coeffs_ref_filtered = coeffs_ref[energy_mask_ref]

energies_s1_filtered = energies_s1[energy_mask_s1]
coeffs_s1_filtered = coeffs_s1[energy_mask_s1]

energies_s2_filtered = energies_s2[energy_mask_s2]
coeffs_s2_filtered = coeffs_s2[energy_mask_s2]

print(f"\nFiltered to energy range {energy_min:.1e} - {energy_max:.1e} eV:")
print(f"Reference: {len(energies_ref_filtered)} points")
print(f"Sample 1: {len(energies_s1_filtered)} points")
print(f"Sample 2: {len(energies_s2_filtered)} points")

# Show some sample values
print(f"\nFirst 5 energy points (eV): {energies_ref_filtered[:5]}")
print(f"Reference coeffs (first 5): {coeffs_ref_filtered[:5]}")
print(f"Sample 1 coeffs (first 5): {coeffs_s1_filtered[:5]}")
print(f"Sample 2 coeffs (first 5): {coeffs_s2_filtered[:5]}")

# Check value ranges
print(f"\nReference coeff range: {coeffs_ref_filtered.min():.6f} to {coeffs_ref_filtered.max():.6f}")
print(f"Sample 1 coeff range: {coeffs_s1_filtered.min():.6f} to {coeffs_s1_filtered.max():.6f}")
print(f"Sample 2 coeff range: {coeffs_s2_filtered.min():.6f} to {coeffs_s2_filtered.max():.6f}")
```

```python
# Extract uncertainty data from the covariance matrix
print("\n" + "="*60)
print("EXTRACTING UNCERTAINTY DATA FROM COVARIANCE MATRIX")
print("="*60)

# Parameters for uncertainty extraction
isotope = 26056  # Fe-56
mt = 2
legendre_order = 1

print(f"Getting uncertainties for isotope {isotope}, MT={mt}, order={legendre_order}")

# Use the dataframe approach - we know it has the right data
print("Extracting from dataframe...")
try:
    df = cov.to_dataframe()
    
    # Get the diagonal element (same L_H and L_V) for our isotope/MT/L combination
    diagonal_row = df[
        (df['ISO_H'] == isotope) & 
        (df['REAC_H'] == mt) & 
        (df['L_H'] == legendre_order) &
        (df['ISO_V'] == isotope) & 
        (df['REAC_V'] == mt) & 
        (df['L_V'] == legendre_order)
    ]
    
    if len(diagonal_row) > 0:
        print("Found diagonal covariance element!")
        
        # Extract the energy grid and uncertainties
        energy_list = diagonal_row['ENE'].iloc[0]
        std_matrix = diagonal_row['STD'].iloc[0]
        
        print(f"Energy list length: {len(energy_list)}")
        print(f"STD matrix type: {type(std_matrix)}")
        print(f"STD matrix length: {len(std_matrix)}")
        print(f"STD matrix[0] length: {len(std_matrix[0]) if isinstance(std_matrix[0], list) else 'Not a list'}")
        
        # The STD should be a matrix where diagonal elements are variances
        uncert_energies = np.array(energy_list)
        
        # Extract diagonal elements and take sqrt to get standard deviations
        if isinstance(std_matrix, list) and len(std_matrix) > 0:
            if isinstance(std_matrix[0], list):
                # It's a 2D matrix - extract diagonal
                n_elements = min(len(std_matrix), len(std_matrix[0]))
                diagonal_variances = np.array([std_matrix[i][i] for i in range(n_elements)])
                uncert_values = np.sqrt(diagonal_variances)
                
                print(f"Matrix size: {len(std_matrix)} x {len(std_matrix[0])}")
                print(f"Using diagonal elements: {n_elements}")
            else:
                # It's already a 1D array of variances
                uncert_values = np.sqrt(np.array(std_matrix))
        else:
            print(f"Unexpected STD format: {type(std_matrix)}")
            uncert_values = None
            
        if uncert_values is not None:
            print(f"Energy array length: {len(uncert_energies)}")
            print(f"Uncertainty array length: {len(uncert_values)}")
            
            # Make sure arrays are the same length
            min_length = min(len(uncert_energies), len(uncert_values))
            uncert_energies = uncert_energies[:min_length]
            uncert_values = uncert_values[:min_length]
            
            # Filter to our energy range
            energy_mask = (uncert_energies >= energy_min) & (uncert_energies <= energy_max)
            uncert_energies_filtered = uncert_energies[energy_mask]
            uncert_values_filtered = uncert_values[energy_mask]
            
            print(f"Total uncertainty points: {len(uncert_energies)}")
            print(f"Filtered uncertainty points: {len(uncert_energies_filtered)}")
            print(f"Energy range: {uncert_energies_filtered.min():.2e} - {uncert_energies_filtered.max():.2e} eV")
            print(f"Uncertainty range: {uncert_values_filtered.min():.6f} - {uncert_values_filtered.max():.6f}")
            print(f"First 5 energies: {uncert_energies_filtered[:5]}")
            print(f"First 5 uncertainties (relative): {uncert_values_filtered[:5]}")
            print(f"Mean uncertainty: {uncert_values_filtered.mean():.6f}")
            
    else:
        print("No diagonal element found!")
        
except Exception as e2:
    print(f"Error with dataframe approach: {e2}")
    import traceback
    traceback.print_exc()
```

```python
# Calculate actual perturbations and compare with uncertainties
print("\n" + "="*60)
print("CALCULATING PERTURBATIONS AND COMPARING WITH UNCERTAINTIES")
print("="*60)

# We need to interpolate everything to a common energy grid
from scipy.interpolate import interp1d

print(f"Reference energy points: {len(energies_ref_filtered)}")
print(f"Sample 1 energy points: {len(energies_s1_filtered)}")
print(f"Sample 2 energy points: {len(energies_s2_filtered)}")
print(f"Uncertainty energy points: {len(uncert_energies_filtered)}")

# Use the reference energy grid as our common grid
common_energies = energies_ref_filtered

# Interpolate sample coefficients to the reference energy grid
interp_s1 = interp1d(energies_s1_filtered, coeffs_s1_filtered, kind='linear', bounds_error=False, fill_value='extrapolate')
interp_s2 = interp1d(energies_s2_filtered, coeffs_s2_filtered, kind='linear', bounds_error=False, fill_value='extrapolate')

coeffs_s1_interp = interp_s1(common_energies)
coeffs_s2_interp = interp_s2(common_energies)

# Interpolate uncertainties to the reference energy grid
uncert_interp = interp1d(uncert_energies_filtered, uncert_values_filtered, kind='linear', bounds_error=False, fill_value='extrapolate')
uncertainties_at_common_energies = uncert_interp(common_energies)

print(f"Common energy grid points: {len(common_energies)}")

# Calculate relative perturbations for both samples
perturbation_s1 = (coeffs_s1_interp - coeffs_ref_filtered) / coeffs_ref_filtered
perturbation_s2 = (coeffs_s2_interp - coeffs_ref_filtered) / coeffs_ref_filtered

print(f"Sample 1 perturbations range: {perturbation_s1.min():.6f} to {perturbation_s1.max():.6f}")
print(f"Sample 2 perturbations range: {perturbation_s2.min():.6f} to {perturbation_s2.max():.6f}")
print(f"Expected uncertainties range: {uncertainties_at_common_energies.min():.6f} to {uncertainties_at_common_energies.max():.6f}")

# Calculate the ratio of actual perturbations to expected uncertainties
ratio_s1 = np.abs(perturbation_s1) / uncertainties_at_common_energies
ratio_s2 = np.abs(perturbation_s2) / uncertainties_at_common_energies

print(f"\nPerturbation-to-uncertainty ratios:")
print(f"Sample 1 mean ratio: {ratio_s1.mean():.2f} (should be ~1 for 1-sigma perturbations)")
print(f"Sample 1 ratio range: {ratio_s1.min():.2f} to {ratio_s1.max():.2f}")
print(f"Sample 2 mean ratio: {ratio_s2.mean():.2f}")
print(f"Sample 2 ratio range: {ratio_s2.min():.2f} to {ratio_s2.max():.2f}")

# Check what percentage of points are within reasonable bounds (1-3 sigma)
within_1sigma_s1 = np.sum(ratio_s1 <= 1.0) / len(ratio_s1) * 100
within_2sigma_s1 = np.sum(ratio_s1 <= 2.0) / len(ratio_s1) * 100
within_3sigma_s1 = np.sum(ratio_s1 <= 3.0) / len(ratio_s1) * 100

within_1sigma_s2 = np.sum(ratio_s2 <= 1.0) / len(ratio_s2) * 100
within_2sigma_s2 = np.sum(ratio_s2 <= 2.0) / len(ratio_s2) * 100
within_3sigma_s2 = np.sum(ratio_s2 <= 3.0) / len(ratio_s2) * 100

print(f"\nPercentage of perturbations within bounds:")
print(f"Sample 1: {within_1sigma_s1:.1f}%  1, {within_2sigma_s1:.1f}%  2, {within_3sigma_s1:.1f}%  3")
print(f"Sample 2: {within_1sigma_s2:.1f}%  1, {within_2sigma_s2:.1f}%  2, {within_3sigma_s2:.1f}%  3")

# Show some specific examples
print(f"\nFirst 10 energy points comparison:")
print("Energy (eV)    | Ref Coeff | S1 Coeff  | S2 Coeff  | S1 Pert % | S2 Pert % | Uncert % | S1/ | S2/")
print("-" * 100)
for i in range(min(10, len(common_energies))):
    e = common_energies[i]
    ref = coeffs_ref_filtered[i]
    s1 = coeffs_s1_interp[i]
    s2 = coeffs_s2_interp[i]
    pert1 = perturbation_s1[i] * 100
    pert2 = perturbation_s2[i] * 100
    unc = uncertainties_at_common_energies[i] * 100
    r1 = ratio_s1[i]
    r2 = ratio_s2[i]
    print(f"{e:11.0f} | {ref:9.6f} | {s1:9.6f} | {s2:9.6f} | {pert1:8.2f} | {pert2:8.2f} | {unc:7.2f} | {r1:4.1f} | {r2:4.1f}")

print(f"\nSUMMARY:")
print(f"- Expected uncertainties (1): {uncertainties_at_common_energies.mean()*100:.2f}%  {uncertainties_at_common_energies.std()*100:.2f}%")
print(f"- Sample 1 mean perturbation: {np.abs(perturbation_s1).mean()*100:.2f}% (ratio: {ratio_s1.mean():.1f})")
print(f"- Sample 2 mean perturbation: {np.abs(perturbation_s2).mean()*100:.2f}% (ratio: {ratio_s2.mean():.1f})")

if ratio_s1.mean() > 3 or ratio_s2.mean() > 3:
    print("  WARNING: Perturbations are much larger than expected uncertainties!")
    print("   This suggests the perturbation factors applied during sampling may be incorrect.")
elif ratio_s1.mean() > 2 or ratio_s2.mean() > 2:
    print("  CAUTION: Perturbations are larger than typical 1-2 range.")
    print("   This might indicate over-perturbation during sampling.")
else:
    print(" Perturbations appear consistent with expected uncertainties.")
```
